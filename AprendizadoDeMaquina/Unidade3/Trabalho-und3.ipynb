Vamos separar o boxplot para cada coluna individualmente para ficar melhor de visualizar

```python
import matplotlib.pyplot as plt
import math

def Customboxplot(df):
  # Número de colunas no DataFrame
  num_columns = len(df.columns)

  # Definir 5 colunas por linha
  num_cols = 5
  # Calcular o número de linhas necessárias
  num_rows = math.ceil(num_columns / num_cols)

  # Criar subplots com o número apropriado de linhas e colunas
  fig, axes = plt.subplots(num_rows, num_cols, figsize=(3*num_cols, 3*num_rows))

  # Flatten caso tenhamos mais de uma linha e coluna de subplots
  axes = axes.flatten()

  # Iterar sobre as colunas e plotar cada boxplot
  for i, column in enumerate(df.columns):
      df.boxplot(column, ax=axes[i])
      axes[i].set_title(column)

  # Remover eixos não utilizados (caso o número de colunas não seja múltiplo de 5)
  for j in range(i + 1, len(axes)):
      fig.delaxes(axes[j])

  # Exibir os gráficos
  plt.tight_layout()
  plt.show()

Customboxplot(df)
```

```python
# Remover outliers

def remove_outliers(df, columns):
    df_: pd.DataFrame = df.copy()
    df_ = df_.astype(float)

    for column in columns:
        q1 = df_[column].quantile(0.25)  # primeiro quartil
        q3 = df_[column].quantile(0.75)  # terceiro quartil
        iqr = q3 - q1

        low_lim = q1 - 1.5 * iqr
        high_lim = q3 + 1.5 * iqr

        # Substitui valores menores que o limite inferior pelo limite inferior
        df_.loc[df_[column] < low_lim, column] = float(low_lim)

        # Substitui valores maiores que o limite superior pelo limite superior
        df_.loc[df_[column] > high_lim, column] = float(high_lim)

    return df_
```

# Holdout

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import optuna

x = df.drop(columns=['Diagnosis'])
y = df['Diagnosis']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# KNN_param_grid = {
#     'n_neighbors': range(1,40,2),
#     'weights':['uniform','distance'],
#     'algorithm':['ball_tree','kd_tree'],
#     'p':[1,2,3],
# }

# KNN_grid = GridSearchCV(knn,KNN_param_grid,verbose = 10)
# KNN_grid.fit(x_train,y_train)

def knn_objective(trial):
    n_neighbors = trial.suggest_int('n_neighbors', 1, 39, step=2) 
    weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
    algorithm = trial.suggest_categorical('algorithm', ['auto', 'ball_tree', 'kd_tree'])
    p = trial.suggest_int('p',1,3)

    model = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, algorithm=algorithm, p=p)

    score = cross_val_score(model, x_train, y_train, cv=5).mean()

    return score
```

```python
knn_study = optuna.create_study(direction='maximize', sampler=optuna.samplers.RandomSampler(seed=42))
```

```python
knn_study.optimize(knn_objective, n_trials=200)
```

```python
# KNN_grid.best_params_
knn_study.best_params
```

```python
# KNN_grid.best_estimator_
knn_study.best_trial
```

```python
# KNN_grid.best_score_
knn_study.best_value
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# KNN_grid_predictions = KNN_grid.predict(x_test)
KNN_model = KNeighborsClassifier(**knn_study.best_params)
KNN_model.fit(x_train, y_train)
KNN_predictions = KNN_model.predict(x_test)

accuracy = accuracy_score(y_test, KNN_predictions)
precision = precision_score(y_test, KNN_predictions)
recall = recall_score(y_test, KNN_predictions)
f1 = f1_score(y_test, KNN_predictions)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1: {f1}')
```

# Árvore de Decisão




```python
from sklearn.tree import DecisionTreeClassifier

# DTR_param_grid = {
#     'criterion': ['gini', 'entropy', 'log_loss'],
#     'splitter': ['best', 'random'],
#     'max_depth': [None, 10, 20, 30],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4],
#     'max_features': [None, 'sqrt', 'log2'],
#     'max_leaf_nodes': [None, 10, 20],
# }

# DTR_grid = GridSearchCV(dtr,DTR_param_grid,verbose = 4)
# DTR_grid.fit(x_train,y_train)

def dtr_objective(trial):
    criterion = trial.suggest_categorical('criterion', ['gini', 'entropy', 'log_loss'])
    splitter = trial.suggest_categorical('splitter', ['best', 'random'])
    max_depth = trial.suggest_categorical('max_depth', [None, 10, 20, 30, 40])
    min_samples_split = trial.suggest_categorical('min_samples_split', [2, 5, 10, 20])
    min_samples_leaf = trial.suggest_categorical('min_samples_leaf', [1, 2, 4, 8])
    max_features = trial.suggest_categorical('max_features', [None, 'sqrt', 'log2'])
    max_leaf_nodes = trial.suggest_categorical('max_leaf_nodes', [None, 10, 20, 30, 40])

    model = DecisionTreeClassifier(criterion=criterion, splitter=splitter, max_depth=max_depth, 
                                  min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf, max_features=max_features,
                                  max_leaf_nodes=max_leaf_nodes)

    score = cross_val_score(model, x_train, y_train, cv=5).mean()

    return score
```

```python
dtr_study = optuna.create_study(direction='maximize', sampler=optuna.samplers.RandomSampler(seed=42))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
models = ['KNN', 'Árvore', 'Random Forest', 'Extra Tree']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']

# Valores para cada modelo (sem tratamento e com tratamento)
values_without_treatment = [
    [0.6041, 0.3559, 0.1071, 0.1647],  # KNN
    [0.9387, 0.9267, 0.9031, 0.9147],  # Árvore
    [0.9665, 0.9588, 0.9490, 0.9538],  # Random Forest
    [0.9349, 0.9497, 0.8673, 0.9067]   # Extra Tree
]

values_with_treatment = [
    [0.7100, 0.6852, 0.3776, 0.4868],  # KNN
    [0.8104, 0.8456, 0.5867, 0.6928],  # Árvore
    [0.8160, 0.8943, 0.5612, 0.6897],  # Random Forest
    [0.8216, 0.9032, 0.5714, 0.7000]   # Extra Tree
]

# Configurações para o gráfico
bar_width = 0.35
index = np.arange(len(models))

# Criar gráfico para cada métrica
plt.figure(figsize=(18, 10))

for i, metric in enumerate(metrics):
    plt.subplot(2, 2, i + 1)  # Cria uma grade de 2x2 para os gráficos
    plt.bar(index, [values_without_treatment[j][i] for j in range(len(models))], bar_width, label='Sem Tratamento', color='lightcoral')
    plt.bar(index + bar_width, [values_with_treatment[j][i] for j in range(len(models))], bar_width, label='Com Tratamento', color='lightskyblue')
    
    # Configurações do gráfico
    plt.xlabel('Modelos')
    plt.ylabel('Valor da Métrica')
    plt.title(f'Comparativo de {metric}')
    plt.xticks(index + bar_width / 2, models)
    plt.legend()

# Ajusta o layout e exibe os gráficos
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados fornecidos com 3 dígitos de precisão
models = ['KNN', 'Árvore de Decisão', 'Random Forest', 'Extra Tree']
Accuracy = [0.604, 0.939, 0.967, 0.935]
Precision = [0.356, 0.927, 0.959, 0.950]
Recall = [0.107, 0.903, 0.949, 0.867]
F1 = [0.165, 0.915, 0.954, 0.907]

# Função para criar um gráfico de barras com cores diferentes
def create_bar_chart(x, y, title, ylabel, ylim, color):
    plt.figure(figsize=(8, 5))
    plt.bar(x, y, color=color)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.ylim(ylim)
    plt.tight_layout()
    plt.show()

# Criar gráficos separados para cada métrica com cores diferentes
create_bar_chart(models, Accuracy, 'Accuracy', 'Accuracy', [0.5, 1.0], 'tab:blue')
create_bar_chart(models, Precision, 'Precision', 'Precision', [0.2, 1.0], 'tab:orange')
create_bar_chart(models, Recall, 'Recall', 'Recall', [0.0, 1.0], 'tab:green')
create_bar_chart(models, F1, 'F1 Score', 'F1 Score', [0.0, 1.0], 'tab:red')
```
